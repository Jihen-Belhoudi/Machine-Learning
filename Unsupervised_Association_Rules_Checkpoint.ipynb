```python
toy_dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],

        ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],

        ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],

        ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],

      ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
```

```python
##Apriori Preparation

import mlxtend
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
te=TransactionEncoder()
te_ary=te.fit(toy_dataset).transform(toy_dataset)    #Apply one-hot-encoding on our dataset
toys=pd.DataFrame(te_ary, columns=te.columns_)  #Creating a new DataFrame from our Numpy array
toys
```

```python
##Support Code
from mlxtend.frequent_patterns import apriori
##Support with column names
frequent_itemsets=apriori(toys, min_support=0.6, use_colnames=True)
#Instead of column indices we can use column names.
frequent_itemsets
```

```python
##Confidence Code
##In case we want to extract rules based on other metrics like confidence,
##we can use association_rules from mlxtend.frequent_patterns library.

from mlxtend.frequent_patterns import association_rules
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.7) # associate items
```

```python
##Lift code
##Associating based on Lift

from mlxtend.frequent_patterns import association_rules
association_rules(frequent_itemsets,metric="lift",min_threshold=1.25)
```

## Read in the Data

```python
# Read in the data
import pandas as pd
Market = pd.read_csv("/content/Market_Basket_Optimisation.csv", header=None)
```

```python
# Convert the DataFrame into a list of lists and dropping nans
transactions = []
for i in range(0,len(Market)):
    transactions.append([str(Market.values[i, j]) for j in range(0, Market.shape[1]) if pd.notna(Market.values[i, j])])
```

```python
# Apriori Preparation
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)  # Apply one-hot-encoding on our dataset
SuperMarket = pd.DataFrame(te_ary, columns=te.columns_)  # Creating a new DataFrame from our Numpy array
```

```python
##Support Code
from mlxtend.frequent_patterns import apriori
##Support with column names
frequent_itemsets=apriori(SuperMarket, min_support=0.05, use_colnames=True)
#Instead of column indices we can use column names.
frequent_itemsets
```

```python
##Confidence Code
from mlxtend.frequent_patterns import association_rules
rules=association_rules(frequent_itemsets,metric="confidence",min_threshold=0.2) # associate items
rules
```

```python
##Lift code
##Associating based on Lift

from mlxtend.frequent_patterns import association_rules
rules=association_rules(frequent_itemsets,metric="lift",min_threshold=1.15)
rules
```
